This code explores the maps and their intensity values.


```python
#import global packages
import numpy as np
import astropy.io.fits as pyfits
from astropy.table import Table
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns

# H-alpha map extension
ext = "HA_FLUX_CORR"
```

```python
# Compile galaxy and sne data
data = Table.read('../Data/1.GroupedSampleData.csv', format='csv')

data[0:1]
```

```python
# Explore H-alpha map and check value ranges

#open H-alpha map
Ha_hdu = pyfits.open(data["maps_nat"][0])
Ha_map = Ha_hdu[ext].data

#set cmap for masking nan values
cmap = mpl.colormaps.get_cmap('viridis')  # viridis is the default colormap for imshow
cmap.set_bad(color='red')

# clean values, removing nans (unobserved edges of map and masked stars)
keep   = np.where(np.isfinite(Ha_map))
map_Ha  = Ha_map[keep]

# clip extreme values (close up stars washing out data)
clip_Ha = np.clip(map_Ha,0,5000)

# plot map and histograms of Halpha intensity values
fig, (ax0, ax1, ax2) = plt.subplots(1, 3)
fig.set_figheight(3)
fig.set_figwidth(9)
ax0.imshow(Ha_map, cmap=cmap)
sns.histplot(map_Ha, bins=10, ax=ax1)
ax1.set_title("All H-alpha Intensity Values")
sns.histplot(clip_Ha, bins=10, ax=ax2)
ax2.set_title("Clipped H-alpha Intensity Values")
fig.tight_layout()
plt.show()
plt.close()

print("H-alpha maps have the filename:", data["maps_nat"][0])
print("and extension #1 ")
print("Most of the H-alpha intensity values are between 0-1000")
print("Max Value: ", np.max(map_Ha))
print("Ave Value: ", np.average(map_Ha))
print("Med Value: ", np.median(map_Ha))
```

```python
# Explore Nebular Mask map and check value ranges

#open Nebular Mask map
HII_hdu = pyfits.open(data["maps_HII"][0])
HII_map = HII_hdu[0].data

#set cmap for masking nan values
cmap = mpl.colormaps.get_cmap('viridis')  # viridis is the default colormap for imshow
cmap.set_bad(color='red')

# clean values, removing nans (unobserved edges of map and masked stars)
keep   = np.where(np.isfinite(HII_map))
map_HII  = HII_map[keep]

# plot map and histogram of nebular mask values
fig, (ax0, ax1) = plt.subplots(1, 2)
fig.set_figheight(3)
fig.set_figwidth(6)
ax0.imshow(HII_map, cmap=cmap)
sns.histplot(map_HII, bins=10, ax=ax1)
ax1.set_title("All Nebular Mask Values")
fig.tight_layout()
plt.show()
plt.close()

print("Non-nebular regions are set = -1")
print("Nebular regions are named and numbered from ~1-1000")
```
